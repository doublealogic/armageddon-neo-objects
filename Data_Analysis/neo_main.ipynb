```python
# Setup and Importing Dependencies
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
```

```python
# Creates DataFrame from cleaned Future NEO CSV file
clean_future_df = pd.read_csv("../CSV_Output/Future_NEO_Data.csv")

# Displays DataFame
clean_future_df
```

```python
# Creates separate DataFrame for all NEOs large
large_future_neos = clean_future_df.loc[clean_future_df["Max Diameter (km)"] > 0.1]

large_future_neos
```

```python
# Variables for Plot
nom_distance = clean_future_df["CA Distance Nominal (km)"]
min_diameter = clean_future_df["Min Diameter (km)"]

# Plot Parameters
plt.scatter(nom_distance, min_diameter, marker="o", facecolor="purple", edgecolor="black")
plt.title(f"Future NEOs Best Case Scenario")
plt.xlabel("Nominal Distance (km)")
plt.ylabel("Min Diameter (km)")
plt.grid(True)

# Saves Plot as a PNG
plt.savefig("../Final_Output/Future_NEO_Best_Case.png")

# Showcases Plot
plt.show()
```

```python
# Variables for Plot
min_distance = clean_future_df["CA Distance Minimum (km)"]
max_diameter = clean_future_df["Max Diameter (km)"]

# Plot Parameters
plt.scatter(min_distance, max_diameter, marker="o", facecolor="green", edgecolor="black")
plt.title(f"Future NEOs Worst Case Scenario")
plt.xlabel("Minimum Distance (km)")
plt.ylabel("Max Diameter (km)")
plt.grid(True)

# Saves Plot as a PNG
plt.savefig("../Final_Output/Future_NEO_Worst_Case.png")

# Showcases Plot
plt.show()
```

```python
# Variables for Plot
lrg_nom_distance = large_future_neos["CA Distance Nominal (km)"]
lrg_min_diameter = large_future_neos["Min Diameter (km)"]

# Plot Parameters
plt.scatter(lrg_nom_distance, lrg_min_diameter, marker="o", facecolor="yellow", edgecolor="black")
plt.title(f"Large Future NEOs Best Case Scenario")
plt.xlabel("Nominal Distance (km)")
plt.ylabel("Min Diameter (km)")
plt.grid(True)

# Saves Plot as a PNG
plt.savefig("../Final_Output/Large_Future_NEO_Best_Case.png")

# Showcases Plot
plt.show()
```

```python
# Creates DataFrame from cleaned Future NEO CSV file
past_neo_df = pd.read_csv("../CSV_Output/Past_NEO_Data.csv")

# Displays DataFame
past_neo_df
```

```python
#Multiply by 1000 and convert ALL maximum diameters to meters to make easier to work with
past_neo_df["Max Diameter (m)"] = past_neo_df["Max Diameter (km)"] * 1000
past_neo_df
```

```python
#Print Max Diameter Max and Min in meters
print(past_neo_df["Max Diameter (m)"].max()) 
print(past_neo_df["Max Diameter (m)"].min())
```

```python
# Create bins in which to place values based upon Max Diameter
bins = [0, 49, 99, 999, 1999]
# Create labels for these bins
group_labels = ["0 to 49 m", "50 to 99 m", "100 to 999m", "1000 to 2000 m"]
```

```python
# Slice the data and place it into bins
pd.cut(past_neo_df["Max Diameter (m)"], bins, labels=group_labels)
past_neo_df["bins"] = pd.cut(past_neo_df["Max Diameter (m)"], bins, labels=group_labels)
```
