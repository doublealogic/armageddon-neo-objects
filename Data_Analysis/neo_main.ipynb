```python
# Setup and Importing Dependencies
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
```

```python
# Creates DataFrame from cleaned Future NEO CSV file
clean_future_df = pd.read_csv("../CSV_Output/Future_NEO_Data.csv")

# Displays DataFame
clean_future_df
```

```python
# Creates separate DataFrame for all NEOs large
large_future_neos = clean_future_df.loc[clean_future_df["Max Diameter (km)"] > 0.1]

large_future_neos
```

```python
# Variables for Plot
nom_distance = clean_future_df["CA Distance Nominal (km)"]
min_diameter = clean_future_df["Min Diameter (km)"]

# Plot Parameters
plt.scatter(nom_distance, min_diameter, marker="o", facecolor="purple", edgecolor="black")
plt.title(f"Future NEOs Best Case Scenario")
plt.xlabel("Nominal Distance (km)")
plt.ylabel("Min Diameter (km)")
plt.grid(True)

# Saves Plot as a PNG
plt.savefig("../Final_Output/Future_NEO_Best_Case.png")

# Showcases Plot
plt.show()
```

```python
# Variables for Plot
min_distance = clean_future_df["CA Distance Minimum (km)"]
max_diameter = clean_future_df["Max Diameter (km)"]

# Plot Parameters
plt.scatter(min_distance, max_diameter, marker="o", facecolor="green", edgecolor="black")
plt.title(f"Future NEOs Worst Case Scenario")
plt.xlabel("Minimum Distance (km)")
plt.ylabel("Max Diameter (km)")
plt.grid(True)

# Saves Plot as a PNG
plt.savefig("../Final_Output/Future_NEO_Worst_Case.png")

# Showcases Plot
plt.show()
```

```python
# Variables for Plot
lrg_nom_distance = large_future_neos["CA Distance Nominal (km)"]
lrg_min_diameter = large_future_neos["Min Diameter (km)"]

# Plot Parameters
plt.scatter(lrg_nom_distance, lrg_min_diameter, marker="o", facecolor="yellow", edgecolor="black")
plt.title(f"Large Future NEOs Best Case Scenario")
plt.xlabel("Nominal Distance (km)")
plt.ylabel("Min Diameter (km)")
plt.grid(True)

# Saves Plot as a PNG
plt.savefig("../Final_Output/Large_Future_NEO_Best_Case.png")

# Showcases Plot
plt.show()
```

```python
# Variables for Plot
lrg_min_distance = large_future_neos["CA Distance Minimum (km)"]
lrg_max_diameter = large_future_neos["Max Diameter (km)"]

# Plot Parameters
plt.scatter(lrg_min_distance, lrg_max_diameter, marker="o", facecolor="red", edgecolor="black")
plt.title(f"Large Future NEOs Worst Case Scenario")
plt.xlabel("Minimum Distance (km)")
plt.ylabel("Max Diameter (km)")
plt.grid(True)

# Saves Plot as a PNG
plt.savefig("../Final_Output/Large_Future_NEO_Worst_Case.png")

# Showcases Plot
plt.show()
```

```python
# Calculates Linear Regression and Sets up Linear Regression Line
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(nom_distance, min_diameter)
regress_values = nom_distance * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot Parameters and Adds Linear Regression Line
plt.scatter(nom_distance, min_diameter)
plt.plot(nom_distance, regress_values, "r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel("Nominal Distance (km)")
plt.ylabel("Min Diameter (km)")
plt.title("Linear Regression of Future NEOs Best Case Scenario")

# Saves Plot as a PNG
plt.savefig("../Final_Output/Regression_Future_NEO_Best_Case.png")

# Calculates and Prints Pearson Coefficient of Plot
print(f"The r-value is: {round(st.pearsonr(nom_distance, min_diameter)[0],2)}")
```

```python
# Creates DataFrame from cleaned Future NEO CSV file
past_neo_df = pd.read_csv("../CSV_Output/Past_NEO_Data.csv")

# Displays DataFame
past_neo_df
```

```python
#Multiply by 1000 and convert ALL maximum diameters to meters to make easier to work with
past_neo_df["Max Diameter (m)"] = past_neo_df["Max Diameter (km)"] * 1000
past_neo_df
```

```python
#Print Max Diameter Max and Min in meters
print(past_neo_df["Max Diameter (m)"].max()) 
print(past_neo_df["Max Diameter (m)"].min())
```

```python
# Create bins in which to place values based upon Max Diameter
bins = [0, 49, 99, 999, 1999]
# Create labels for these bins
group_labels = ["0 to 49 m", "50 to 99 m", "100 to 999m", "1000 to 2000 m"]
```

```python
# Slice the data and place it into bins
past_neo_df['Diameter Grps'] = pd.cut(past_neo_df['Max Diameter (m)'], bins, labels=group_labels)
df_testing = past_neo_df.groupby('Diameter Grps').count()
```

```python
df_testing = past_neo_df.copy()
df_testing.set_index('Diameter Grps', inplace = True)
```
